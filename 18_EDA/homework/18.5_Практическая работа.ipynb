1. Загрузите датасет и ознакомьтесь с количеством пропущенных значений.

```python
import pandas as pd
import missingno as msno
import numpy as np
df = pd.read_csv("data/vehicles_dataset.csv")
```

```python
#процент пропущенных значений в каждом признаке:
missing_val = ((df.isna().sum()/len(df))*100).round(2).sort_values()
missing_val
```

2. Обработайте пропуски в годе выпуска автомобиля.

```python
#так как пустых ячеек в year меньше 1% , то можно удалить эти строки
df_clean = df.copy()
#df_clean["year"].describe()

missing_val2 = ((df_clean.isna().sum()/len(df_clean))*100).round(2).sort_values()


#inplace = True ничего не возвращает(null). Он изменяет существующий фрейм данных с помощью желаемой операции 
#и делает это «на месте», то есть в исходном фрейме данных.
df_clean.dropna(inplace=True, subset=['year'])
df_clean.shape
```

3. Обработайте пропуски в типе топлива.

```python
#так как пустых ячеек в fuel меньше 1% , то можно удалить эти строки
df_clean.dropna(inplace=True, subset=['fuel'])
df_clean.shape
```

4. Обработайте пропуски в модели автомобиля.

```python
#так как пустых ячеек в fuel около 1% , то можно удалить эти строки
df_clean.dropna(inplace=True, subset=['model'])
df_clean.shape
```

5. Обработайте пропуски в статусе автомобиля.

```python
#так как пустых ячеек в title_status около 1% , то можно удалить эти строки
df_clean.dropna(inplace=True, subset=['title_status'])
df_clean.shape
df_clean.title_status.value_counts()
```

6. Преобразуйте типы данных в годе выпуска автомобиля.

```python
df_types = df_clean.copy()
df_types.year = df_types.year.astype(int)
df_types.info()
```

7. Преобразуйте типы данных в категориальные переменные.

```python
#под этим заданием понимается , что нужно дать всем колонкам тип object??? какие типы...
df_types = df_types.astype("object")
df_types.info()
```

8. Определите выбросы в годе выпуска.

```python
df_out = df_types.copy()

def calculate_outliers(data):
    q25 = data.quantile(0.25)
    q75 = data.quantile(0.75)
    iqr = q75 - q25
    boundaries = (q25-1.5*iqr,q75+1.5*iqr)
    return boundaries

boundaries = calculate_outliers(df_out.year)

#находим значения ниже нижней границы и выше верхней
is_outlier = (df_out.year<boundaries[0])|(df_out.year>boundaries[1])
#доля выбросов 3.75%
is_outlier.sum()/len(df_out)
df_out[is_outlier]
```
