По умолчанию функция размещает элементы в порядке возрастания. Например, `sorted([2, 4, 1, 0])` вернёт `[0, 1, 2, 4]`. Подробнее про работу функции можно почитать в [**официальной документации**](https://docs.python.org/3/library/functions.html?highlight=sorted#sorted)

````python
##
## Вставьте свой код сюда.
##

list_of_tuples = []
with open('/home/rootocop/Документы/Data Science/курсы/5.Command line interface/words_freq.txt') as file:
    
    for line in file:
        freq, word = line.strip().split(" ")
        list_of_tuples.append((int(freq),word))
    

a = sorted(list_of_tuples,reverse=True)
list_top_10 = a[:10]
print("ТОП 10 самых популярных слов:\n")
#print(a)
for i in list_top_10:
    
    print ("Частота слова:",i[1],"=",i[0],)
        
        
        
````

Сортировать словарь по значениям сложнее, чем список объектов, которые можно сравнивать непосредственно. Обратите внимание на параметр key функции [**sorted**](https://docs.python.org/3/library/functions.html?highlight=sorted#sorted).

````python
##
## Вставьте свой код сюда.
##
new_dict = {}
list_first_lat = []
with open('/home/rootocop/Документы/Data Science/курсы/5.Command line interface/words_freq.txt') as file:

#делаю список с количеством слов и заглавными буквами  

    for line in file:
        key, value = line.strip().split(" ")
        list_first_lat.append((value[0], int(key)))
#в наш пустой словарь складываю одинаковые заглавные "буквы", которые повторяюся. Получаю к-во всех уник. символов.       
    for key, val in list_first_lat:
        
        new_dict[key] = new_dict.get(key, 0) + val

max_value = max(new_dict.values())

#возможен такой случай что наибольшее значение может повториться(max) по этому поможет словарь(генератор словаря):

final_dict = {k: v for k, v in new_dict.items() if v == max_value}

#просить первый элемент словаря не совсем умно, но когда у меня в словаре -один- элемент, можно пренебречь)
#делов том что при каждом новом вызове элемента словаря , мы можем не получить этот элемент по списку - то есть не итерируемый
#короче, в словаре позиции элементов не закреплены и не пронумерованы что бы так делать...

print(f"Самая популярная буква: {list(final_dict.keys())[0]} \nколичество букв составляет: {list(final_dict.values())[0]}")                
                
        
#прошу куратора дать комментарий , как правильно сделать вывод ответа на задание. Спасибо!        
        

     


    
````

###Что нужно сделать
Напишите функцию `circ_sqrt`. Она должна принимать один параметр `r` типа `float` и возвращать квадратный корень из длины окружности радиуса `r`.

### Подсказки
- Используйте функции встроенного модуля math.
- В работе функции учтите область определения квадратного корня.

````python
from math import sqrt, pi

def circ_sqrt(r):
    
    
    if type(r) == float and r >= 0:

        sqr_cir = sqrt(2*pi*r)

    else: 
        print ("Число должно быть вещественным из области определения квадратного корня, введите еще раз...")
    
    return sqr_cir

print(circ_sqrt(3.0))
````

````python
list_of_tuples = []
dict_of_lens = {}
with open('/home/rootocop/Документы/Data Science/курсы/5.Command line interface/anna_words.txt') as file:

    for line in file:
        word = line.strip()
        len_word = len(word)
        list_of_tuples.append((word,len_word))
        
    for key, val in list_of_tuples:
        
        dict_of_lens[val] = dict_of_lens.get(val, 0) + 1
        
    len_sorted_keys = sorted(dict_of_lens) 
    
    final_dict = {}
    for i in len_sorted_keys:
        final_dict[i] = dict_of_lens[i]
        
print(final_dict)
````
